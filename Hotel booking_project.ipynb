#  Data Analysis and Visualisations

# Total amount of reservation cancellation in data

```python
cancelled_per = df['is_canceled'].value_counts(normalize=True)
cancelled_per

# around 62% reservation are not cancelled , around 37% reservation are cancelled
```

```python
plt.figure(figsize=(7,4))
plt.title('Reservation Status Count',fontsize=15)
plt.bar(['not cancelled','cancelled'],df['is_canceled'].value_counts(),color='green',alpha=0.87,width=0.6,edgecolor='black')
plt.gcf().set_facecolor('lightgrey')
plt.show()

```

# Depending on the Hotels to see ratio of cancellation 

```python
plt.figure(figsize=(10,5))
sns.set_style("darkgrid")
ax=sns.countplot(x='hotel',hue='is_canceled',data=df,palette='nipy_spectral',edgecolor='black')
plt.title('Reservation Status in Different Hotel',size=16,fontweight='bold')
plt.xlabel('hotel',size=15,color="purple",fontweight='bold')
plt.ylabel('no. of reservation',size=15,color="purple",fontweight='bold')
plt.legend(['not cancelled','cancelled'],fontsize=12)
sns.set_context("paper")
for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

## Here's to check the Percentage of reservation cancelled & not cancelled in both  Resort Hotel and City Hotel......

```python
resort_hotel=df[df['hotel']== 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel=df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
# mean value of average daily rate of hotels
# to check price affected on cancellation of both the Hotels
```

```python
resort_hotel= resort_hotel.groupby('reservation_status_date')[['adr']].mean()
resort_hotel
```

```python
city_hotel= city_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel
```

# To check= City Hotel  price is less than Resort Hotel for some days

```python
plt.figure(figsize=(15,6))
sns.set_style("whitegrid")
plt.plot(resort_hotel.index,resort_hotel['adr'], label='resort_hotel')
plt.plot(city_hotel.index,city_hotel['adr'], label='city_hotel')
plt.title('Average Daily rate in City and Resort Hotel',size=20,fontweight='bold',color='black')
plt.xlabel('Date', size=18,color="red",fontweight='bold')
plt.ylabel('Average Daily Rate (ADR)', size=17,color="red",fontweight='bold')
legend=plt.legend(fontsize=18)

for text in legend.get_texts():
    text.set_color("red") 
    
plt.show() 
```

```python
df['month']= df['reservation_status_date'].dt.month
```

```python
plt.figure(figsize=(12,5))
sns.set_style("ticks")
sns.countplot(x='month',hue='is_canceled',data=df,palette='autumn')
plt.title('Reservation Status Per Month',size=19,color='green',fontweight='bold')
plt.xlabel('month',size=15,color="black",fontweight='bold')
plt.ylabel('number of reservations',size=15,color="black",fontweight='bold')
plt.legend(['not cancelled','cancelled'],fontsize=12)
plt.gcf().set_facecolor('lightgrey')
plt.show()
```

```python
top_10 = df['country'].value_counts().nlargest(10)

plt.figure(figsize=(10, 7))
plt.title('Top 10 Countries with Reservation Cancellations',fontsize=16)
plt.pie(top_10, autopct='%.2f%%', labels=top_10.index)

plt.show()
```
